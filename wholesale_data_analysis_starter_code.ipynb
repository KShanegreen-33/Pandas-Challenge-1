Import the data and use Pandas to learn more about the dataset.

```python
import pandas as pd

def currency_format_millions(amount):
    """
    Converts a dollar amount to millions and formats it with a dollar sign and 'M' for millions.
    
    Args:
    amount (float or int): The dollar amount to be converted and formatted.
    
    Returns:
    str: The formatted dollar amount in millions with a dollar sign.
    """
    millions = round(amount / 1_000_000, 2)  # Convert to millions
    return f"${millions}M"  # Format as currency with 'M' for millions

df = pd.read_csv('client_dataset.csv')
df.head()
```

## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']

#Valadating that the subtotal was created
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
# Assuming unit_weight is also a column
df['total_weight'] = df['unit_weight'] * df['qty']

# Applying the shipping price logic
df['shipping_price'] = df['total_weight'].apply(lambda x: x * 7 if x > 50 else x * 10)

# Display the updated DataFrame
df[['client_id', 'qty', 'unit_weight', 'total_weight', 'shipping_price']]


```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
# Calculate the total price and round it to 2 decimal places

df['total_price'] = ((df['subtotal'] + df['shipping_price']) * 1.0925).round(2)

# Display the updated DataFrame with total price in currency format
df[['client_id', 'qty', 'unit_price', 'subtotal', 'shipping_price', 'total_price']]

```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
# Create the cost column by multiplying unit_cost by qty and adding the shipping_price
df['cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']

# Display the updated DataFrame
df[['client_id', 'qty', 'unit_cost', 'shipping_price', 'cost']]

```

```python
# Create a column for the profit of each line using line cost and line price
# Create the profit column by subtracting cost from total price
df['profit'] = df['total_price'] - df['cost']

# Display the updated DataFrame with profit
df[['client_id', 'qty', 'unit_price', 'total_price', 'cost', 'profit']]

```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
# Note: the instructions are a little confusing here because when I built the top client list above it was based on the five clients that had the most entries in the data as instructed, not by the quanity spend. I used the user list from part 1 here.
top_clients_data = df[df['client_id'].isin(top_client_ids)]
client_order_totals = top_clients_data.groupby('client_id')['total_price'].sum().reset_index()
client_order_totals['total_price'] = client_order_totals['total_price'].apply(lambda x: f"${x:,.2f}")
print(client_order_totals[['client_id', 'total_price']])

```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
summary_df = top_clients_data.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('total_price', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()

print(summary_df)
```

```python
# Format the data and rename the columns to names suitable for presentation.

summary_df.rename(columns={
    'client_id': 'Client ID',
    'total_units_purchased': 'Units Purchased',
    'total_shipping_price': 'Total Shipping Price',
    'total_revenue': 'Total Revenue',
    'total_profit': 'Total Profit'
}, inplace=True)

# Apply the currency_format_millions function to the money columns only
money_columns = ['Total Shipping Price', 'Total Revenue', 'Total Profit']
for col in money_columns:
    summary_df[col] = summary_df[col].apply(currency_format_millions)

# Display the formatted summary DataFrame
print(summary_df)

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_summary_df = summary_df.sort_values(by='Total Profit', ascending=False, key=lambda x: x.str.replace('$','').str.replace('M','').astype(float))

# Display the sorted DataFrame
print(sorted_summary_df)
```
